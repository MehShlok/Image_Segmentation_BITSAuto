```python
import cv2
import os

class Extractor:
    path = None
    dir = None
    fps = None
    interval = None
    total_frames = None

    def __init__(self, path, dir):
        Extractor.path = path
        Extractor.dir = dir
        Extractor.cap = Extractor.load_video(path)
        Extractor.fps = Extractor.cap.get(cv2.CAP_PROP_FPS)
        Extractor.interval = int(Extractor.fps * 5)
        Extractor.total_frames = int(Extractor.cap.get(cv2.CAP_PROP_FRAME_COUNT))


    def load_video(path):
        return cv2.VideoCapture(path)


    def save_frame(dir, frame, count):
        if not os.path.exists(dir):
            os.makedirs(dir)
        frame_filename = os.path.join(dir, f'frame_{count:05d}.png')
        cv2.imwrite(frame_filename, frame)
        print(f"success {frame_filename}")

    @staticmethod
    def extract_frames():
        frame_count = 0
        i = 0
        while i < Extractor.total_frames:
            Extractor.cap.set(cv2.CAP_PROP_POS_FRAMES, i)
            ret, frame = Extractor.cap.read()
            if not ret:
                break
            Extractor.save_frame(Extractor.dir, frame, frame_count)
            frame_count += 1
            i += Extractor.interval


# sample main function to be used with the above class imported

# def main():
#     path = '/content/20240831_145516.mp4'
#     dir = 'frames2'

#     extractor = Extractor(path, dir)

#     extractor.extract_frames()


#     extractor.cap.release()
#     cv2.destroyAllWindows()



#     print("Extraction complete!")

# if __name__ == "__main__":
#     main()
```
