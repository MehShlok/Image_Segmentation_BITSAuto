```python
import os
import cv2
import numpy as np
from pycocotools.coco import COCO
from segment_anything import SamPredictor, sam_model_registry

# Define paths
train_folder = r"C:\Users\abbas\Desktop\2-1\Image-Segementation\unet_project\images"
output_folder = r"C:\Users\abbas\Desktop\2-1\Image-Segementation\unet_project\masks"  # Define where to save the masks
os.makedirs(output_folder, exist_ok=True)  # Create output directory if it doesn't exist
coco_json_path = os.path.join(train_folder, "_annotations.coco.json")

# Load COCO annotations
coco = COCO(coco_json_path)

# Initialize SAM model
sam = sam_model_registry["vit_h"](checkpoint=r"C:\Users\abbas\Desktop\2-1\Image-Segementation\unet_project\segment_anything\checkpoints\sam_vit_h_4b8939.pth")
predictor = SamPredictor(sam)

# Iterate over the images in the dataset
for img_id in coco.getImgIds():
    img_info = coco.loadImgs([img_id])[0]
    img_path = os.path.join(train_folder, img_info['file_name'])
    
    # Load the image
    image = cv2.imread(img_path)
    predictor.set_image(image)

    # Get annotations (polygons) for this image
    ann_ids = coco.getAnnIds(imgIds=[img_id])
    anns = coco.loadAnns(ann_ids)

    # Create a colored mask for overlay (3 channels)
    colored_mask = np.zeros_like(image)

    for ann in anns:
        # Get the segmentation polygon points
        polygons = ann['segmentation']

        # Check if polygons are in the expected format
        if isinstance(polygons, list) and all(isinstance(p, list) for p in polygons):
            # Prepare point coordinates for SAM
            point_coords = []
            point_labels = []  # Initialize point_labels

            for polygon in polygons:
                # Flatten the polygon coordinates into point coordinates
                for i in range(0, len(polygon), 2):
                    point_coords.append((polygon[i], polygon[i + 1]))
                    point_labels.append(1)  # Set label for all points to 1 (for "Roads")

                # Draw the original polygon on the image (in red)
                polygon_points = np.array(polygon).reshape((-1, 1, 2)).astype(np.int32)
                cv2.polylines(image, [polygon_points], isClosed=True, color=(0, 0, 255), thickness=2)

            # Convert to numpy arrays for SAM
            point_coords = np.array(point_coords)
            point_labels = np.array(point_labels)

            # Pass point coordinates and labels to SAM for segmentation
            masks, scores, logits = predictor.predict(point_coords=point_coords, point_labels=point_labels)

            # Combine masks
            combined_mask = np.zeros_like(masks[0])  # Assuming all masks have the same shape
            for mask in masks:
                combined_mask = np.logical_or(combined_mask, mask)  # Combine masks

            # Fill the colored mask for the areas identified by SAM
            colored_mask[combined_mask > 0] = [0, 255, 0]  # Green color for the mask
            
            # Calculate predicted IoU for the current mask
            predicted_iou = np.mean([mask.mean() for mask in masks])  # Average IoU for all masks
            predicted_iou_text = f"Accuracy={predicted_iou:.2f}"  # Format text
            
            # Display predicted IoU on the image
            text_position = (image.shape[1] - 200, 30)  # Top right corner
            cv2.putText(image, predicted_iou_text, text_position, 
                        cv2.FONT_HERSHEY_SIMPLEX, 0.6, (0, 0, 0), 2, cv2.LINE_AA)  # Black text

    # Overlay the mask on the original image
    overlayed_image = cv2.addWeighted(image, 0.7, colored_mask, 0.3, 0)

    # Save the overlayed image with original polygons and IoU
    overlayed_output_path = os.path.join(output_folder, f"overlayed_with_polygons_{img_info['file_name']}")
    cv2.imwrite(overlayed_output_path, overlayed_image)
    print(f"Saved overlayed image to {overlayed_output_path}")

# Cleanup
cv2.destroyAllWindows()
```

```python
import numpy as np
import cv2
from pycocotools.coco import COCO

# Assuming you have already loaded your image, annotation, and prediction as in your code above
def calculate_iou(pred_mask, true_mask):
    intersection = np.logical_and(pred_mask, true_mask).sum()
    union = np.logical_or(pred_mask, true_mask).sum()
    return intersection / union if union != 0 else 0

def calculate_dsc(pred_mask, true_mask):
    intersection = np.logical_and(pred_mask, true_mask).sum()
    return 2 * intersection / (pred_mask.sum() + true_mask.sum()) if (pred_mask.sum() + true_mask.sum()) != 0 else 0

# Assuming the rest of your code (SAM model, COCO annotations) is loaded
# Initialize variables to track metrics
total_iou = 0
total_dsc = 0
num_images = 0

# Iterate over the images in the dataset
for img_id in coco.getImgIds():
    img_info = coco.loadImgs([img_id])[0]
    img_path = os.path.join(train_folder, img_info['file_name'])
    
    # Load the image
    image = cv2.imread(img_path)
    predictor.set_image(image)

    # Get annotations (polygons) for this image
    ann_ids = coco.getAnnIds(imgIds=[img_id])
    anns = coco.loadAnns(ann_ids)

    # Create the ground truth mask
    true_mask = np.zeros_like(image[:, :, 0])  # Single channel mask
    for ann in anns:
        polygons = ann['segmentation']
        for polygon in polygons:
            polygon_points = np.array(polygon).reshape((-1, 1, 2)).astype(np.int32)
            cv2.fillPoly(true_mask, [polygon_points], 1)

    # Initialize a variable for the predicted mask
    colored_mask = np.zeros_like(image[:, :, 0])  # Single channel mask

    for ann in anns:
        polygons = ann['segmentation']
        point_coords = []
        point_labels = []  # Initialize point_labels

        for polygon in polygons:
            for i in range(0, len(polygon), 2):
                point_coords.append((polygon[i], polygon[i + 1]))
                point_labels.append(1)  # Set label for all points to 1

            # Draw the original polygon on the image (in red)
            polygon_points = np.array(polygon).reshape((-1, 1, 2)).astype(np.int32)
            cv2.polylines(image, [polygon_points], isClosed=True, color=(0, 0, 255), thickness=2)

        point_coords = np.array(point_coords)
        point_labels = np.array(point_labels)

        # Pass point coordinates and labels to SAM for segmentation
        masks, _, _ = predictor.predict(point_coords=point_coords, point_labels=point_labels)

        # Combine masks
        combined_mask = np.zeros_like(masks[0])  # Assuming all masks have the same shape
        for mask in masks:
            combined_mask = np.logical_or(combined_mask, mask)  # Combine masks

        # Calculate IoU and DSC
        iou = calculate_iou(combined_mask, true_mask)
        dsc = calculate_dsc(combined_mask, true_mask)

        # Accumulate for average calculation
        total_iou += iou
        total_dsc += dsc
        num_images += 1

# Calculate average IoU and DSC after all images are processed
average_iou = total_iou / num_images
average_dsc = total_dsc / num_images

print(f"Average IoU: {average_iou:.4f}, Average DSC: {average_dsc:.4f}")
```
